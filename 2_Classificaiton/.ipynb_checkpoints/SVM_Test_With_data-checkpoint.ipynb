```python
lead_time = data_df["lead_time"]
prev_cancels = data_df["previous_cancellations"]
print("lead_time\n")
print(lead_time)
print("\npre_cancellation\n")
print(prev_cancels)
```

```python
hotel_data = [[l,p] for l,p in zip(lead_time, prev_cancels)]
print(hotel_data)
```

```python
is_canceled = data_df["is_canceled"]
print(is_canceled)
```

```python
# from sklearn.model_selection import train_test_split
# x_train, x_test, y_train, y_test = train_test_split(hotel_data, is_canceled, shuffle=True, test_size=0.2, random_state=20)
x_train = hotel_data[0:int(len(hotel_data)*0.8)]
x_test = hotel_data[int(len(hotel_data)*0.8):int(len(hotel_data))]

y_train = is_canceled[0:int(len(is_canceled)*0.8)]
y_test = is_canceled[int(len(is_canceled)*0.8):int(len(is_canceled))]

print(len(y_train))
print(len(y_test))
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', C=0.01, gamma=0.5)
classifier.fit(x_train, y_train) 
```

```python
classifier.score(x_train,y_train)
```

```python
classifier.score(x_test,y_test)
```

```python
result = classifier.predict(x_test)
prediction = np.asarray(result, dtype = 'int')
actual = np.asarray(y_test, dtype = 'int')
```

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(actual, prediction)
print(confusion_matrix)
```
